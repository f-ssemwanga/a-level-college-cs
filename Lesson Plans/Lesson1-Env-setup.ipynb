# Year 12 Computer Science - Lesson Plan Overview

**Duration:** 2 hours (2 lessons)
**Repository:** `f-ssemwanga/a-level-college-cs`

---

## Lesson Structure

### Lesson 1 (60 minutes): Environment Setup & GCSE Task
1. **Tool Installation** (25 min) - Git Bash, Python, VS Code, Extensions
2. **GCSE Coding Challenge** (30 min) - Password Strength Checker 
3. **Setup Verification** (5 min)

### Lesson 2 (60 minutes): Jupyter & A-Level Task
1. **Jupyter Server Setup** (15 min) - Install and configure Jupyter
2. **GitHub & Repository Setup** (15 min) - Clone repo, basic Git workflow
3. **A-Level Coding Challenge** (25 min) - Student Grade Management System
4. **Git Workflow Practice** (5 min) - Commit and push changes

---

## Repository Structure

Students will clone and work with:
```
a-level-college-cs/
├── setup-guides/
│   ├── 01-vscode-setup.ipynb
│   ├── 02-jupyter-notebook-fix.ipynb
├── challenges/
│   ├── password-checker-challenge.md
│   ├── grade-manager-challenge.ipynb
│   ├── grade-manager-hints.ipynb
├── teacher-solutions/
│   ├── password_checker_solution.py
│   ├── grade_manager_solution.ipynb
```

---

### GCSE Challenge (Lesson 1)
- **Students create:** `password_checker.py` file
- **Tool focus:** VS Code + Code Runner
- **Skill level:** Bridge from GCSE to A-level thinking

### A-Level Challenge (Lesson 2) 
- **Format:** Jupyter notebook with markdown cells containing challenges
- **Students add:** Code cells with their solutions
- **Tool focus:** Jupyter notebooks + Git workflow
- **Skill level:** Functions, parameters, data structures
- **Choice:** Students can use dictionaries, lists, or 2D arrays
